```python
import numpy as np
from scipy import stats
import scipy
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.proportion import proportions_ztest
```

```python
# slide #67 paired t-test
#df1, df2를 직접 입력
df1 = pd.DataFrame({'before':[720,589,780,648,720,589,780,648,780,648]})
df2 = pd.DataFrame({'after':[810,670,790,712,810,670,790,712,790,712]})
#paired t-test
t_result = stats.ttest_rel(df1,df2)
#paired t-test값 p값 저장
t,p = t_result.statistic.round(3),t_result.pvalue.round(3)
#paired t-test값 p값 출력
print("Paired t-test")
print("t검정통계량: {}".format(t))
print("p-value: {}".format(p))
```

### 결론: 유의수준 5%에서 검정결과 P값이 0.000이므로 학원 수강 전후의 점수는 차이가 있다고 할 수 있다.

```python
# slide #68 paired t-test 실습
df1 = pd.DataFrame({'before':[720,589,780,648,720,589,780,648,780,648]})
df2 = pd.DataFrame({'after':[710,580,787,712,750,600,782,670,790,680]})
t_result = stats.ttest_rel(df1,df2)
t,p = t_result.statistic.round(3),t_result.pvalue.round(4)
print("Paired t-test")
print("t검정통계량: {}".format(t))
print("p-value: {}".format(p))
```

### 결론: 유의수준 5%에서 검정한 결과 p-value가 0.0497이므로 귀무가설을 기각할 수 없다. 그래서 학원 수강 후의 성적에 차이가 있다고 할 수 없다.

```python
# slide #69 ANOVA 연습
df1 = pd.DataFrame({'A':[892,623,721,678,723,790,720,670,690,771]})
df2 = pd.DataFrame({'B':[721,821,910,678,723,790,711,790,745,891]})
df3 = pd.DataFrame({'C':[621,915,888,721,894,834,841,912,845,889]})
#ANOVA 실행
f_result = stats.f_oneway(df1,df2,df3)
#ANOVA의 f값 p값 저장
f,p = f_result.statistic.round(3),f_result.pvalue.round(3)
#ANOVA의 f값 p값 출력
print("One-way ANOVA")
print("F검정통계량: {}".format(f))
print("p-value: {}".format(p))
```

```python
# slide #70 1 Proportion test 연습
from statsmodels.stats.proportion import proportions_ztest
#변수 선언
count = 15
nobs = 100
value = .1
#Proportion test 실행
stat,pval = proportions_ztest(count,nobs,value)
#Proportion test 결과 출력
print("1 Proportion test")
print("p검정통계량: {0:0.3f}".format(stat))
print("p-value: {0:0.3f}".format(pval))
```

### P-value가 0.161로 귀무가설을 기각할 수 없다. 그러므로 품질개선 결과로 기존보다 전체 고객 중 만족도 차이가 있다고 할 수 없다.

```python
from statsmodels.stats.proportion import proportions_ztest
#변수 선언
count = 10
nobs = 100
value = .2
#Proportion test 실행
stat,pval = proportions_ztest(count,nobs,value)
#Proportion test 결과 출력
print("1 Proportion test")
print("p검정통계량: {0:0.3f}".format(stat))
print("p-value: {0:0.3f}".format(pval))
```

### P-value가 0.001로 귀무가설을 기각한다. 그러므로 품질개선 결과로 기존보다 전체 고객 중 만족도 차이가 있다.

```python
# slide #72 2 Proportion test 연습
#변수 선언
count = np.array([4,1])
nobs = np.array([1000,1200])
#Proportion test 실행
stat,pval = proportions_ztest(count,nobs)
#Proportion test 결과 출력
print("2 Proportion test")
print("p검정통계량: {0:0.3f}".format(stat))
print("p-value: {0:0.3f}".format(pval))
```

### H0 : p1(불량률을 의미) = p2, H1 : p1 != p2 p-value가 0.120으로 유의수준보다 크므로 귀무가설을 기각할 수 없다. 
### 그러므로 두 공정의 불량률은 차이가 있다고 할 수 없다.

```python
# slide #73 2 Proportion test 연습
#변수 선언
count = np.array([14,5])
nobs = np.array([1200,1200])
#Proportion test 실행
stat,pval = proportions_ztest(count,nobs)
#Proportion test 결과 출력
print("2 Proportion test")
print("p검정통계량: {0:0.3f}".format(stat))
print("p-value: {0:0.3f}".format(pval))
```

### H0 : p1(불량률을 의미) = p2, H1 : p1 != p2 p-value가 0.038으로 유의수준보다 작으므므로 귀무가설을 기각한다. 
### 그러므로 두 공정의 불량률은 차이가 있다.

```python
df = pd.DataFrame({'HSG':[270,260,236,234],'SS':[228,285,225,262],'SPA':[277,284,231,208]})
#chi-square test 실행
chi,pval,dof,expected = stats.chi2_contingency(df.T)
#chi-square test 결과 출력
print("chi-square test")
print("chisq:{0:0.3f}".format(chi))
print("p:{0:0.3f}".format(pval))
print('degree of freedom: {}'.format(dof))
print('expected value:\n{}'.format(expected.round(3)))
```

### 결론: 근무조별로 제품을 생산하는 부하 차이가 있다고 할 수 있다.

```python
df = pd.DataFrame({'AS':[18,8,4,4,3,3],'가격':[1,2,1,1,1,25],'성능':[8,14,3,2,3,8],'확장성':[7,5,4,3,1,10],'디자인':[10,5,9,2,1,2],'안정성':[9,9,5,7,1,1],'기능성':[10,4,4,3,1,7]})
#chi-square test 실행
chi,pval,dof,expected = stats.chi2_contingency(df.T)
#chi-square test 결과 출력
print("chi-square test")
print("chisq:{0:0.3f}".format(chi))
print("p:{0:0.3f}".format(pval))
print('degree of freedom: {}'.format(dof))
print('expected value:\n{}'.format(expected.round(3)))
```
